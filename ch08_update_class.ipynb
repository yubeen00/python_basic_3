# 8장 데이터 준비하기 : 조인, 병합, 변형

## * DataFrame 합성
* 판다스는 두 개 이상의 데이터프레임을 하나로 합치는 데이터 병합(merge)이나 연결(concatenate)을 지원

## * merge 함수를 사용한 데이터프레임 병합
* merge 함수는 두 데이터 프레임의 공통 열 혹은 인덱스를 기준으로 두 개의 테이블을 합친다.
* 이 때 기준이 되는 열, 행의 데이터를 키(key)라고 한다.

```python
# ex1 
df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": pd.Series(range(7), dtype="Int64")})

df2 = pd.DataFrame({"key": ["a", "b", "d"],
                    "data2": pd.Series(range(3), dtype="Int64")})
print(df1)
print()
print(df2)
```

```python
pd.merge(df1, df2, how="inner")
```

```python
pd.merge(df1, df2, on="key", how="inner")
```

```python
# ex2 - 공통된 열 이름이 없는 경우
df3 = pd.DataFrame({"lkey": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": pd.Series(range(7), dtype="Int64")})
df4 = pd.DataFrame({"rkey": ["a", "b", "d"],
                    "data2": pd.Series(range(3), dtype="Int64")})
print(df3, "\n\n", df4)
pd.merge(df3, df4, how = "inner", left_on="lkey", right_on="rkey")
```

```python
# ex3 : 외부 조인
pd.merge(df1, df2, how="outer")
```

```python
# 외부 조인 : DataFrame의 색인과 일치하지 않는 왼쪽 또는 오른쪽 DataFrame 객체의 해을 Na으로 처리
f3 = pd.DataFrame({"lkey": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": pd.Series(range(7), dtype="Int64")})
df4 = pd.DataFrame({"rkey": ["a", "b", "d"],
                    "data2": pd.Series(range(3), dtype="Int64")})
print(df3, "\n\n", df4)
pd.merge(df3, df4, left_on = "lkey", right_on = "rkey", how="outer")
```
